## Feature Engineering and Handling of Missing Values

```python
#Dropping Unnnecessary columns
df=df_fifa.drop(columns=['ID','Photo','Flag','Club Logo'])
```

```python
#Converting wage into float
lst=[]
for x in df['Wage'].values:
    lst.append(float(str(x)[str(x).index('€')+1:-1]+'000'))
df['Wage']=lst
```

```python
#Converting value column into float
lst=[]
for x in df['Value'].values:
    if x[-1]=='M':
        lst.append(float(x[x.index('€')+1:x.index('M')])*1000000)
    elif x[-1]=='K':
        lst.append(float(x[x.index('€')+1:x.index('K')])*1000)
    else:
        lst.append(float(x[x.index('€')+1:]))
df['Value']=lst
```

```python
df[['Day','Year']]=df.Joined.str.split(', ',expand=True)
```

```python
df.Year.unique()
```

```python
year=['2004', '2018', '2017', '2011', '2015', '2012', '2014', '2005',
       '2010', '2016', '2008', '2013','2007', '2009', '2002',
       '2003', '2006', '2001', '1991', '1998', '2000', '1999']
```

```python
df.Year.mode()
```

```python
def year_set(x):
    if x in year:
        return x
    else:
        return '2018'
    
df['Year']=df.Year.apply(year_set)
```

```python
#Converting year into int
lst=[]
for x in df.Year.values:
    lst.append(int(x.strip()))
df['Year']=lst
```

```python
#dropping more unnecessary columns
df.drop(columns=['Day','Joined','Loaned From'],inplace=True)
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
attributes=['Crossing','Finishing','HeadingAccuracy','ShortPassing', 'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']
```

```python
#Checking general distribution of Numerical attributes
rows=17
cols=2
counter=1
plt.rcParams['figure.figsize']=[15,8]
for i in attributes:
    plt.subplot(rows,cols,counter)
    sns.distplot(df.loc[df.Overall==62,i])
    counter+=1
    
plt.tight_layout
plt.show()
```

```python
for i in attributes:
    df[i]=df[i].fillna(round(df.loc[df.Overall==62,i].median()))
#taking overall 62 because all missing attribute rows had overall 62
```

```python
Category_attribute=['Preferred Foot', 'International Reputation', 'Weak Foot',
       'Skill Moves', 'Work Rate', 'Body Type']
```

```python
for i in Category_attribute:
    df[i]=df[i].fillna(df[i].mode().values[0])
```

```python
#Handling missing values for Jersey Number
df.Position=df.Position.fillna('Not Known')
df['Jersey Number']=df['Jersey Number'].fillna('Unknown')
```

```python
#Handling Missing Values for Club
df.Club=df.Club.fillna('Not Known')
```

```python
#Handling missing values for Contract Valid Until
df['Contract Valid Until']=df['Contract Valid Until'].fillna(df['Contract Valid Until'].mode().values[0])
```

```python
df['Contract Valid Until'].unique()
```

```python
valid_until_years=['2021', '2022', '2020', '2023', '2019', '2024','2025', '2026','2018']
```

```python
#Keeping only years from Contract Vaalid until
def valid_fill(x):
    if x in valid_until_years:
        return int(x.strip())
    else:
        return int(x[x.index(',')+2:].strip())

df['Contract Valid Until']=df['Contract Valid Until'].apply(valid_fill)
```

```python
df.Weight.unique()
```

```python
weights=['159lbs', '183lbs', '150lbs', '168lbs', '154lbs', '163lbs',
       '146lbs', '190lbs', '181lbs', '192lbs', '176lbs', '172lbs',
       '148lbs', '165lbs', '196lbs', '161lbs', '187lbs', '212lbs',
       '170lbs', '203lbs', '157lbs', '185lbs', '130lbs', '174lbs',
       '207lbs', '134lbs', '141lbs', '152lbs', '179lbs', '132lbs',
       '201lbs', '198lbs', '209lbs', '214lbs', '143lbs', '205lbs',
       '137lbs', '194lbs', '216lbs', '139lbs', '220lbs', '126lbs',
       '218lbs', '123lbs', '227lbs', '128lbs', '223lbs', '225lbs',
       '121lbs', '115lbs', '117lbs', '236lbs', '229lbs', '243lbs',
       '110lbs', '119lbs', '234lbs']
```

```python
#Converting weight into float
def weight_fill(x):
    if x in weights:
        return float(x[:x.index('l')])

df['Weight']=df.Weight.apply(weight_fill)
```

```python
df.Height.unique()
```

```python
heights=["5'7", "6'2", "5'9", "6'4", "5'11", "5'8", "6'0", "5'6", "5'10",
       "6'6", "6'1", "5'4", "6'3", "5'5", "6'5", "6'7", "5'3", "5'2",
       "6'8", "5'1", "6'9"]
```

```python
#Converting Height into cms
def height_fill(x):
    if x in heights:
        return float(x[0])*30.48 + float(x[2:])*2.54
    
df.Height=df.Height.apply(height_fill)
```

```python
df.Height
```

```python
df.columns
```

```python
#Handling Missing Values for Height
df.Height=df.Height.fillna(df.Height.median())
```

```python
#Handling Missing Values for Weight
df.Weight=df.Weight.fillna(df.Weight.median())
```

```python
df['Release Clause']
```

```python
def release_clause(x):
    if x is not None:
        if str(x)[-1]=='M':
            return float(str(x)[str(x).index('€')+1:str(x).index('M')])*1000000
        elif str(x)[-1]=='K':
            return float(str(x)[str(x).index('€')+1:str(x).index('K')])*1000
        
df['Release Clause']=df['Release Clause'].apply(release_clause)    
```

```python
df.loc[(df['Release Clause'].isnull()) & (df.Overall>80),'Release Clause']=df.loc[df.Overall>80,'Release Clause'].median()
df.loc[(df['Release Clause'].isnull()) & (df.Overall>70),'Release Clause']=df.loc[df.Overall>70,'Release Clause'].median()
df.loc[(df['Release Clause'].isnull()) & (df.Overall>60),'Release Clause']=df.loc[df.Overall>60,'Release Clause'].median()
df.loc[(df['Release Clause'].isnull()) & (df.Overall>50),'Release Clause']=df.loc[df.Overall>50,'Release Clause'].median()
df.loc[(df['Release Clause'].isnull()) & (df.Overall>40),'Release Clause']=df.loc[df.Overall>40,'Release Clause'].median()
```
